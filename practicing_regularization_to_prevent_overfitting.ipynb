```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sonar_dataset.csv', header = None)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
for columns in df :
  print(f'{columns} : {df[columns].unique()} ')
```

```python
df.dtypes
```

```python
# as seen above that the last column which is the target column is categorical variable , lets handle it

df =  pd.get_dummies(df, '60', drop_first= True, dtype = int)
df.head()
```

```python
df['60_R'].unique()
```

```python
df.dtypes
```

```python
X = df.drop('60_R', axis= 1)
X.head()
```

```python
y = df['60_R']
y.head()
```

```python
y.value_counts()
```

#first split my data into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X, y , test_size = 0.2, random_state= 1234, stratify = y  )
```

```python
(X_train.shape, X_test.shape)
```

#Lets build my Artificial Neural Networks (ANN) without a dropout layer

I will intentionally build build this model to overfit so I can see if drop_out regularization works

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(1000, input_shape = (60,), activation = 'relu'),
    keras.layers.Dense (1, activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)


model.fit(X_train, y_train, epochs = 100, batch_size = 8)

```

#Evaluation and predicition using the model on new dataset

```python
model.evaluate(X_test, y_test)
```

```python
y_preds = np.round(model.predict(X_test))
y_preds[:5]
```

the model performed not too poorly on the test set , I guess ? Do you guess the same as me?

```python
from sklearn.metrics import classification_report , confusion_matrix

print(classification_report(y_test, y_preds))

```

#Now, Building the ANN model with a dropout layer to see if i can get a better performance


```python
X_train.shape
```

```python
model1 = keras.Sequential([
    keras.layers.Dense(10000, input_shape = (60,), activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation = 'sigmoid')
])

model1.compile(
    optimizer ='adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model1.fit(X_train, y_train, epochs = 100, batch_size = 8)
```

```python
model1.evaluate(X_test, y_test)
```

```python
y_preds = np.round(model1.predict(X_test))
y_preds[:9]
```

```python
from sklearn.metrics import classification_report , confusion_matrix

print(classification_report(y_test, y_preds))
```

```python
#Well , there wasn't any much difference , it seems that the f1-score of the model without dropout performed well , but the class 1 seems to do better  in this case in the recall
```
