```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

```

```python
df  = pd.read_csv('/content/Churn_Modelling.csv')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
#I need to drop the columns that are not needed

df1 = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1 )
df1.head()
```

```python
df1.dtypes
```

```python
df1['Gender'].unique()
```

```python
df1['Gender'] = df1['Gender'].replace({'Female' : 0, 'Male' : 1})
```

```python
df1 = pd.get_dummies(df1, 'Geography', drop_first = True, dtype= int)
df1.head()
```

```python
#realized some columns have unscaled values
sx = MinMaxScaler()

cols_to_scale = ['CreditScore', 'Age', 'Tenure', 'Balance','NumOfProducts','EstimatedSalary']

df1[cols_to_scale] = sx.fit_transform(df[cols_to_scale])
df1[cols_to_scale] .head()
```

```python
df1.head()
```

#Checking if the dataset is balance or not

```python
df['Exited'].value_counts()
```

#it is grossly imbalanced , so we are oversampling the minority class in this case. :

```python
class_0 = df1[df1['Exited'] == 0]
class_1 = df1[df1['Exited'] == 1]
```

```python
(count_class_0,  count_class_1)= df1['Exited'].value_counts()
```

```python
df_sample_0_over = class_0.sample(count_class_1)
df_sample_0_over.head()
```

```python
df_sample_0_over.shape
```

```python
df2 = pd.concat([df_sample_0_over, class_1 ], axis = 0)
df2.head()
```

```python
df2['Exited'].value_counts()
```

```python
X = df2.drop('Exited', axis= 1)
X.head()
```

```python
y = df2['Exited']
y.sample(6)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train , y_test  = train_test_split (X, y , test_size = 0.2 , random_state = 2345)
```
