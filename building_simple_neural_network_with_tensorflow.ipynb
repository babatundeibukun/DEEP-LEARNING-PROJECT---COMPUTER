```python
import numpy as np
import tensorflow as tf
import pandas as pd
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection  import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age', 'affordibility']], df.bought_insurance, test_size= 0.2 , random_state=234)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train.head()
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age']/100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age']/100
```

```python
X_train_scaled.head()
```

#Building a simple neural network with tensorflow

```python
model = keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,), activation='sigmoid', kernel_initializer='ones', bias_initializer= 'zeros')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_scaled, y_train, epochs= 5000)
```

To know how the model performs on an unseen data

```python
model.evaluate(X_test_scaled, y_test)
```

Now, lets predict

```python
model.predict(X_test_scaled)
```
