#Building a logistic regression model

#We have an imbalance dataset


```python
#we cant just throw all of this into our model
##Lets solve the problem of imbalance
```

#Method 1 : oversampling the minority class


```python
#Make a train set into a dataframe for easy manipulation
```

```python
df2 = X_train
df2['churn'] = y_train
```

```python
df2.shape
```

```python

class_0 = df2[df2['churn']== 0]
```

```python
class_1 = df2[df2['churn']== 1]
```

```python
count_0, count_1 = class_0[0], class_1[0]
count_0, count_1
```

#Lets oversample the minority class


```python
df_over_sample =  class_1.sample(count_0, replace= True)
df_over_sample.shape
```

```python
df_train = pd.concat([df_over_sample, class_0],  axis = 0)
df_train.shape
```

```python
df_train['churn'].value_counts()
```

#Now we have balanced dataset

```python
X_train = df_train.drop('churn', axis = 1)
X_train
```

```python
y_train = df_train['churn']
y_train.head()
```

#lets throw the x_train and y_train


```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
y_pred = np.round(model.predict(X_test))
y_pred
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
