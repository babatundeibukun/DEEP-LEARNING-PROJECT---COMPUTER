#As seen above , many of the the columns are categorical columns , lets handle that

```python
#lets zoom in on the categorical columns first
def all_columns_unique_values():
  for cols in df:
    if df[cols].dtypes == 'object' :
      print(f'{cols} : {df[cols].unique()}')

all_columns_unique_values()

```
